```python
import rdflib
import networkx as nx
import ipycytoscape

# Namespaces
SKOS = rdflib.Namespace("http://www.w3.org/2004/02/skos/core#")
RDFS = rdflib.namespace.RDFS
XSD = rdflib.namespace.XSD
OWL = rdflib.namespace.OWL
DBP = rdflib.Namespace("https://dbpedia.org/page/")
DBPP = rdflib.Namespace("https://dbpedia.org/property/")
BSDD = rdflib.Namespace("https://identifier.buildingsmart.org/")
FCL = rdflib.Namespace("https://www.freeclass.de/")

# Create a new RDFLib graph and parse the Turtle file
g = rdflib.Graph()
g.bind("skos", SKOS)
g.bind("rdfs", RDFS)
g.bind("xsd", XSD)
g.bind("owl", OWL)
g.bind("dbp", DBP)
g.bind("dbp", DBPP)
g.bind("bSDD", BSDD)
g.bind("fcl", FCL)
g.parse(r"C:\2024_B4_IsabelleSeifert_RDF_Concrete.ttl")  # Update with your file path

# Function to get label
def get_label(node):
    label = g.value(node, RDFS.label)
    return str(label) if label else str(node)

# Function to get label with prefix
def get_label_with_prefix(node):
    for prefix, uri in g.namespaces():
        if str(node).startswith(uri):
            return f"{prefix}:{node.split(uri)[-1]}"
    return str(node)  # Return full URI if no namespace match

# Print node labels with prefixes
for subj, _, _ in g:
    print(get_label_with_prefix(subj))

# Convert RDFLib graph to a NetworkX graph
nx_graph = nx.DiGraph()

# Function to get edge label with prefix
def get_edge_label(pred):
    for prefix, uri in g.namespaces():
        if str(pred).startswith(uri):
            return f"{prefix}:{pred.split(uri)[-1]}"
    return str(pred)  # Return full URI if no namespace match

for subj, pred, obj in g:
    subj_label = get_label(subj)
    obj_label = get_label(obj)
    
    if not nx_graph.has_node(subj):
        nx_graph.add_node(subj, label=get_label_with_prefix(subj))
    if not nx_graph.has_node(obj):
        nx_graph.add_node(obj, label=get_label_with_prefix(obj))
    
    edge_label = get_edge_label(pred)
    nx_graph.add_edge(subj, obj, label=edge_label)

# Define layout options for concentric layout
layout_options = {
    'name': 'concentric',  # Change layout to concentric
    'animate': True,
    'spacingFactor': 5,  # Adjust spacing between rings of nodes
}

# Convert NetworkX graph to Cytoscape JSON format
cyto_graph = ipycytoscape.CytoscapeWidget(layout=layout_options)
cyto_graph.graph.add_graph_from_networkx(nx_graph, directed=True)

# Define style for the nodes and edges
cyto_graph.set_style([
    {
        'selector': 'node',
        'style': {
            'label': 'data(label)',
            'color': '#000000',
            'text-valign': 'center',
            'text-halign': 'right',
            'background-color': '#11479e'
        }
    },
    {
        'selector': 'edge',
        'style': {
            'label': 'data(label)',
            'curve-style': 'bezier',
            'target-arrow-shape': 'triangle',
            'color': '#9dbaea',
            'line-color': '#9dbaea',
            'target-arrow-color': '#9dbaea'
        }
    }
])

cyto_graph


```

```python
# Export the graph structure to GraphML
nx.write_graphml(nx_graph, "Concrete.graphml")

```
